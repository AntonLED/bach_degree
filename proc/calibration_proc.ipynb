```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("../data/calibration/data_with_fixed_xy.txt", skip_header=2)

fxy_top_x = data[0::2, 1]
fxy_top_y = data[0::2, 2]
fxy_top_z = data[0::2, 3]
fxy_top_Ex_ref = data[0::2, 4]
fxy_top_Ey_ref = data[0::2, 5]
fxy_top_Ez_ref = data[0::2, 6]
fxy_top_Ewx = data[0::2, 7]
fxy_top_Ewy = data[0::2, 8]
fxy_top_Ewz = data[0::2, 9]
fxy_bot_x = data[1::2, 1]
fxy_bot_y = data[1::2, 2]
fxy_bot_z = data[1::2, 3]
fxy_bot_Ex_ref = data[1::2, 4]
fxy_bot_Ey_ref = data[1::2, 5]
fxy_bot_Ez_ref = data[1::2, 6]
fxy_bot_Ewx = data[1::2, 7]
fxy_bot_Ewy = data[1::2, 8]
fxy_bot_Ewz = data[1::2, 9]

data = np.genfromtxt("../data/calibration/data_with_fixed_yz.txt", skip_header=2)

fyz_top_x = data[0::2, 1]
fyz_top_y = data[0::2, 2]
fyz_top_z = data[0::2, 3]
fyz_top_Ex_ref = data[0::2, 4]
fyz_top_Ey_ref = data[0::2, 5]
fyz_top_Ez_ref = data[0::2, 6]
fyz_top_Ewx = data[0::2, 7]
fyz_top_Ewy = data[0::2, 8]
fyz_top_Ewz = data[0::2, 9]
fyz_bot_x = data[1::2, 1]
fyz_bot_y = data[1::2, 2]
fyz_bot_z = data[1::2, 3]
fyz_bot_Ex_ref = data[1::2, 4]
fyz_bot_Ey_ref = data[1::2, 5]
fyz_bot_Ez_ref = data[1::2, 6]
fyz_bot_Ewx = data[1::2, 7]
fyz_bot_Ewy = data[1::2, 8]
fyz_bot_Ewz = data[1::2, 9]

data = np.genfromtxt("../data/calibration/data_with_fixed_xz.txt", skip_header=2)

fxz_top_x = data[0::2, 1]
fxz_top_y = data[0::2, 2]
fxz_top_z = data[0::2, 3]
fxz_top_Ex_ref = data[0::2, 4]
fxz_top_Ey_ref = data[0::2, 5]
fxz_top_Ez_ref = data[0::2, 6]
fxz_top_Ewx = data[0::2, 7]
fxz_top_Ewy = data[0::2, 8]
fxz_top_Ewz = data[0::2, 9]
fxz_bot_x = data[1::2, 1]
fxz_bot_y = data[1::2, 2]
fxz_bot_z = data[1::2, 3]
fxz_bot_Ex_ref = data[1::2, 4]
fxz_bot_Ey_ref = data[1::2, 5]
fxz_bot_Ez_ref = data[1::2, 6]
fxz_bot_Ewx = data[1::2, 7]
fxz_bot_Ewy = data[1::2, 8]
fxz_bot_Ewz = data[1::2, 9]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 5))
fig.suptitle("Fixed xy")

axes[0][0].plot(fxy_bot_z, fxy_top_Ez_ref, ".")
axes[0][0].set_title("Top particle")
axes[0][0].set_xlabel("z")
axes[0][0].set_ylabel("E_z_ref")
axes[0][0].legend()

axes[1][0].plot(fxy_bot_z, fxy_top_Ewz, ".")
axes[1][0].set_title("Top particle")
axes[1][0].set_xlabel("z")
axes[1][0].set_ylabel("E_z_analit")
axes[1][0].legend()

axes[0][1].plot(fxy_bot_z, fxy_bot_Ez_ref, ".")
axes[0][1].set_title("Bot particle")
axes[0][1].set_xlabel("z")
axes[0][1].set_ylabel("E_z_ref")
axes[0][1].legend()

axes[1][1].plot(fxy_bot_z, fxy_bot_Ewz, ".")
axes[1][1].set_title("Bot particle")
axes[1][1].set_xlabel("z")
axes[1][1].set_ylabel("E_z_analit")
axes[1][1].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 5))
fig.suptitle("Fixed xz")

axes[0][0].plot(fxz_bot_y, fxz_top_Ey_ref, ".")
axes[0][0].set_title("Top particle")
axes[0][0].set_xlabel("y")
axes[0][0].set_ylabel("E_y_ref")
axes[0][0].legend()

axes[1][0].plot(fxz_bot_y, fxz_top_Ewy, ".")
axes[1][0].set_title("Top particle")
axes[1][0].set_xlabel("y")
axes[1][0].set_ylabel("E_y_analit")
axes[1][0].legend()

axes[0][1].plot(fxz_bot_y, fxz_bot_Ey_ref, ".")
axes[0][1].set_title("Bot particle")
axes[0][1].set_xlabel("y")
axes[0][1].set_ylabel("E_y_ref")
axes[0][1].legend()

axes[1][1].plot(fxz_bot_y, fxz_bot_Ewy, ".")
axes[1][1].set_title("Bot particle")
axes[1][1].set_xlabel("y")
axes[1][1].set_ylabel("E_y_analit")
axes[1][1].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 5))
fig.suptitle("Fixed yz")

axes[0][0].plot(fyz_bot_x, fyz_top_Ex_ref, ".")
axes[0][0].set_title("Top particle")
axes[0][0].set_xlabel("x")
axes[0][0].set_ylabel("E_x_ref")
axes[0][0].legend()

axes[1][0].plot(fyz_bot_x, fyz_top_Ewx, ".")
axes[1][0].set_title("Top particle")
axes[1][0].set_xlabel("x")
axes[1][0].set_ylabel("E_x_analit")
axes[1][0].legend()

axes[0][1].plot(fyz_bot_x, fyz_bot_Ex_ref, ".")
axes[0][1].set_title("Bot particle")
axes[0][1].set_xlabel("x")
axes[0][1].set_ylabel("E_x_ref")
axes[0][1].legend()

axes[1][1].plot(fyz_bot_x, fyz_bot_Ewx, ".")
axes[1][1].set_title("Bot particle")
axes[1][1].set_xlabel("x")
axes[1][1].set_ylabel("E_x_analit")
axes[1][1].legend()

plt.tight_layout()
plt.show()
```
